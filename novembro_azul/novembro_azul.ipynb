## Rescalling dos dados

```python
# Dividindo em subset de entreda e saída

array = df.values

x = array[:,1:6]
Y = array[:,0]
```

```python
# Alterando a escala dos dados para 0 e 1
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler(feature_range=(0,1))
X = scaler.fit_transform(x)
```

```python
X
```

Estamos preparados para a divisão dos dados de treino e teste

```python
from sklearn.model_selection import train_test_split

# 70% treino, 30% teste
Xtrain,Xtest,Ytrain,Ytest = train_test_split(X,Y,test_size=0.3,random_state=7)
```

Representando a divisão proporcial dos dados

```python
fig = plt.figure(figsize=(10,6))

# valores para grafico
x_graph = ['X', 'Y']
test = [len(Xtest), len(Ytest)] 
train = [len(Xtrain), len(Ytrain)]

#gráfico
plt.bar(x_graph, test,
        width=0.4,
       color=light_slate_gray) 

plt.bar(x_graph, train,
        bottom=test,
       width=0.4,
       color = royal_blue)

plt.legend(["Dados teste", "Dados treino"])
plt.show() 
```

## Modelos Preditivos 

### Random Forest

```python
from sklearn.ensemble import RandomForestClassifier


randomf = RandomForestClassifier(n_estimators=200)
randomf.fit(Xtrain,Ytrain)
pred_randomf = randomf.predict(Xtest)



# Avaliando as previsões
accuracy = accuracy_score(Ytest, pred_randomf)
print("Acurácia: %.3f%%" % (accuracy * 100.0))
confusion_matrix(Ytest,pred_randomf)
```

### Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier

dtree = DecisionTreeClassifier()
dtree.fit(Xtrain,Ytrain)
pred_dtree = dtree.predict(Xtest)



# Avaliando as previsões
accuracy = accuracy_score(Ytest, pred_dtree)
print("Acurácia: %.3f%%" % (accuracy * 100.0))
confusion_matrix(Ytest,pred_dtree)
```

###  KNN

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier()
knn.fit(Xtrain,Ytrain)
pred_knn = knn.predict(Xtest)



# Avaliando as previsões
accuracy = accuracy_score(Ytest, pred_knn)
print("Acurácia: %.3f%%" % (accuracy * 100.0))
confusion_matrix(Ytest,pred_knn)
```

### Logistic Regression

```python
from sklearn.linear_model import LogisticRegression

logreg = LogisticRegression()
logreg.fit(Xtrain,Ytrain)
pred_logreg = logreg.predict(Xtest)


# Avaliando as previsões
accuracy = accuracy_score(Ytest, pred_logreg)
print("Acurácia: %.3f%%" % (accuracy * 100.0))
confusion_matrix(Ytest,pred_logreg)
```
